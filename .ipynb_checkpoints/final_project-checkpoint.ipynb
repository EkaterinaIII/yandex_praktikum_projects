# Телеком — задача проекта

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

**Описание услуг:**

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

**Описание данных:**

Данные состоят из файлов, полученных из разных источников:

- `contract_new.csv` — информация о договоре;
- `personal_new.csv` — персональные данные клиента;
- `internet_new.csv` — информация об интернет-услугах;
- `phone_new.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

# Цель проекта:
Разработать модель, способную своевременно выявлять клиентов, близких к расторжению договора о предоставлении услуг

## Исследование моделей

Напишем небольшую функцию, для расчета целевой метрики в одну строку:

```python
def auc_proba(features_train, target_train,model): 
    predict_proba=model.predict_proba(features_train)
    predictions=roc_auc_score(target_train, predict_proba[:,1])
    return predictions
```

###  LogisticRegression
Чтобы использовать логистичесую регрессию, необходимо проводить масштабирование числовых и кодирование категориальных признаков. Мы будем использовать крос-валидацию, поэтому, используем  Pipeline для которого создадим трансформер OneHotEncoder:

```python
ob_col=list(features.select_dtypes(include=['object']).columns)
try:           
    ohe = OneHotEncoder(handle_unknown='ignore',drop='first', sparse_output=False) #отличия в синтаксисе sparse в разных версиях
except:
    ohe = OneHotEncoder(handle_unknown='ignore',drop='first', sparse =False)

tr_ohe = ColumnTransformer(transformers =[('enc', ohe, ob_col),], remainder ='passthrough')
```

```python
model=LogisticRegression(solver='liblinear',max_iter=1000000)

pipe_lg = Pipeline(steps =[
    ('tr_ohe', tr_ohe),
    ('minmax', MinMaxScaler()),
    ('model', model)
])
```

```python
cvs = cross_val_score(pipe_lg, features_train, target_train, cv = 10,scoring=make_scorer(roc_auc_score))
```

```python
print("ROC AUC: ", cvs.max())
```

Значение метрики - низкое.

### LGBMClassifier

В LGBMClassifier есть встроенный кодировщик, но, чтобы он заработал, необходимо поменять тип данных на 'category':

```python
ob_col=list(features.select_dtypes(include=['object']).columns)

for c in ob_col:
    features_train[c] = features_train[c].astype('category')
    features_test[c] = features_test[c].astype('category')
```

```python
params={            'n_estimators': [28,50,63,100], 
                     'max_depth': [6,7,8,9,11],
                     'learning_rate':[0.3]
                        }

m_lgb = lgb.LGBMClassifier(objective="binary",random_state=random_state)  

grid_l = RandomizedSearchCV(m_lgb, params, cv=4,  verbose=1,random_state=random_state,n_jobs=-1,scoring='roc_auc', n_iter=20 )


begin_l=time.time()
grid_l.fit(features_train, target_train)
end_l=time.time()
```

```python
begin_pred_l=time.time()
predictions= grid_l.best_estimator_.predict(features_train)
end_pred_l=time.time()
time_pred_l=round((end_pred_l-begin_pred_l),4)

print('AUC_valid:',grid_l.best_score_)
print('Accuracy_train:',acc(predictions, target_train))
print('-'*57)
print('Время обучения:',round(end_l-begin_l,2))
print('Скорость предсказания:',time_pred_l)
print('Модель:   LGBMClassifier')
print('Параметры модели:', grid_l.best_params_)
```

### CatBoostClassifier

##  Финансовый анализ

Попробуем посчитать, примерные издержки и принесенную прибыль, работы нашей модели.
Для простоты будем считать, что потенциально отточным клиентам предлагают скидку - 10% от месячного тарифа. Также, сколько-то отточных клиентов модель упустит, и они уйдут. Это все и будут наши издержки

Также, преположим, что люди, которым предложили скидку, остаются хотя бы на месяц. Посчитаем среднемесячную плату, для расчета прибыльности.

Скорее всего, из тех, кто получит скидку, останутся не все, и какая-то часть уйдет. Будем считать, что 20% клиентов уйдет, не смотря на скидку. Поэтому гипотетическую прибыль будем умножать на коэффициент "Полезности скидки"

```python
koef=0.8 #коэффициент "Полезность скидки"
```

```python
month_mean=round(features_test['MonthlyCharges'].mean())
month_mean #среднемесячная плата, в условных единицах
```

```python
discount=0.1*month_mean
discount  #скидка, в условных единицах
```

### Confusion_matrix

Чтобы узнать, количество "упущенных" отточных клиентов и количество "лишних" выданных скидок, удобно использовать Confusion_matrix

```python
predicted_test = grid_l.best_estimator_.predict(features_test)
cm=confusion_matrix(target_test,predicted_test)
```

```python
pl=sns.heatmap(cm, cmap="Spectral", annot= True, fmt="d", annot_kws={"size":13});
pl.axes.set_title("Матрица ошибок", fontsize=16)
pl.set_xlabel("Предсказания", fontsize=14)
pl.set_ylabel("Ответы", fontsize=14);
```

```python
false_neg=cm[1][0] # сколько потенциално отточных модель пропустила
income_no=koef*false_neg*month_mean
print('Потенциальный доход на сумму:',income_no) 
```

```python
over_discount=cm[0][1]*discount
print('Издержки/ Выдали лишней скидки на сумму:',over_discount)
```

```python
true_positive=cm[1][1] # верно выдали скидку
income=koef*true_positive*month_mean
print('Доход:',income)
```

```python
#Прибыль как  доход - издержки-недополученный доход
profit=income-over_discount-income_no
profit
```

```python
#Прибыль, %
profit_per=round(profit/(month_mean*cm.sum())*100,2)
profit_per
```

### Изменим порог

Замечено, что мы много недополучаем средств, т.к. модель находит не всех отточных клиентов. На скидку мы тратим, не так много, поэтому попробуем уменьшить порог, чтобы увеличить полноту и минимизировать сумму (Издержки+Недополученный доход)

```python
probabilities_test = grid_l.best_estimator_.predict_proba(features_test)
probabilities_one_t = probabilities_test[:, 1]

for threshold in np.arange(0, 0.5, 0.02):
    predicted_test = probabilities_one_t > threshold
   
    cm=confusion_matrix(target_test,predicted_test)
    false_neg=cm[1][0]  #сколько потенциално отточных модель пропустила
    income_no=koef*false_neg*month_mean
    over_discount=cm[0][1]*discount #Выдали лишней скидки на сумму
    sum_i=income_no+over_discount #суммарные издержки
    
    print("Порог = {:.2f} | Издержки+недополученный доход = {:.2f} |  (Упустили клиентов = {:.2f},\
Лишняя скидка = {:.2f})".format(threshold,sum_i, income_no,over_discount))
```

Выявили, что суммарные издержки минимальны, при пороге равном 0.08. Пересчитаем прибыль: 

```python
threshold=0.08
predicted_test = probabilities_one_t > threshold
cm=confusion_matrix(target_test,predicted_test)
false_neg=cm[1][0]  #сколько потенциально отточных модель пропустила
income_no=koef*false_neg*month_mean
over_discount=cm[0][1]*discount #Выдали лишней скидки на сумму
sum_i=income_no+over_discount #суммарные издержки
true_positive=cm[1][1] # верно выдали скидку
income=koef*true_positive*month_mean
print('Доход:',round(income,2))
profit=income-over_discount-income_no
print('Прибыль:',round(profit,2))
profit_per=round(profit/(month_mean*cm.sum())*100,2)
print('Прибыль,%:',profit_per)
```

Таким образом, сильно уменьшив порог, мы значительно увеличили полноту предсказания отточных клиентов и увеличили прибыльность нашей модели почти в 5 раз!
